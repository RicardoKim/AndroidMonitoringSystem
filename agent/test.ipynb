```python
import subprocess
from collections import defaultdict
```

```python
adb_process = subprocess.Popen(['adb', 'shell', 'top', '-m', '10'], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
```

```python
section = ""
last_log = ""
log_category = defaultdict(list)
while True:
    log = adb_process.stdout.readline().decode().strip()
    if log != last_log:
        last_section = section
        prefix = log
        if "Task" in prefix :
            section = "Task"
            log_category[section].append(log)
        elif "Mem" in prefix:
            section = "Memory"
            log_category[section].append(log)
        elif "Swap" in prefix:
            section = "Swap"
            log_category[section].append(log)
        elif "PID" in prefix :
            section = "Process"
        elif last_section == "Process" :
            log_category[section].append(log)
        else :
            pass
        last_log = log

    print(log_category['Process'])
    
```

```python
# Initialize variables to store different sections of the log
task_info = ""
memory_info = ""
process_info = ""

# Define a flag to track the current section
current_section = None

# Define the separator line
separator = "---"

# Process each line of the log
for line in log.split('\n'):
    if line.startswith("Tasks:") or line.startswith("Mem:") or line.startswith("Swap:"):
        # Update the current section and reset the data for other sections
        current_section = line
        if current_section == "Tasks:":
            task_info = ""
        elif current_section == "Mem:":
            memory_info = ""
        elif current_section == "Swap:":
            memory_info += "\n" + separator  # Add separator before Swap section
    else:
        # Append the line to the appropriate section
        if current_section == "Tasks:":
            task_info += line + "\n"
        elif current_section == "Mem:":
            memory_info += line + "\n"
        elif current_section == "Swap:":
            memory_info += line + "\n"
        else:
            process_info += line + "\n"

# Print or save the separated sections
print("Task Information:")
print(task_info)
print(separator)
print("Memory Information:")
print(memory_info)
print(separator)
print("Process Information:")
print(process_info)
```
